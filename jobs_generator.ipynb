```python
import itertools

def generate_combinations():
    # Generate all combinations of 7 bits
    bit_combinations = itertools.product([0, 1], repeat=7)
    
    # Iterate over each combination and add '1' at the end
    for bits in bit_combinations:
        yield bits + (1,)

norms = []
# Iterate over generated combinations and print them
for combination in generate_combinations():
    if combination[0]==combination[1] and combination[2]==combination[3] and combination[4]==combination[5] and combination[6]==combination[7]:
        continue
    else:
        norms.append(",".join(str(element) for element in combination))
```
