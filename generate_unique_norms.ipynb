```python
def invert(d: list):
    return [1-d[4], 1-d[5], 1-d[6], 1-d[7], 1-d[0], 1-d[1], 1-d[2], 1-d[3]]

def invert_emotion(d):
    return [d[1], d[0], d[3], d[2], d[5], d[4], d[7], d[6]]

all_norms = []

# Generate all
for i in range(256):
    d = [int(bit) for bit in format(i, '08b')]
    all_norms.append(d)
 
# remove symmetrics
for d in all_norms:
    if invert(d) in all_norms:
        print("Inversion of", d, ":", invert(d), "already in array, so removing it")
        all_norms.remove(invert(d))
print("Total norms without symmetrics:", len(all_norms))
```

```python
# create list of self-symmetrics
self_symmetrics = []
for i in range(256):
    d = [int(bit) for bit in format(i, '08b')]
    if d == invert(d):
        print("Self-symmetric:", d)
        self_symmetrics.append(d)
print("\tThere are", len(self_symmetrics), "self symmetric ebsn's")

# add the self-symmetrics to the list if they aren't there already
for ss in self_symmetrics:
    all_norms.append(ss)

print("Total norms without symmetrics but with self-symmetrics:", len(all_norms))
```

```python
# remove non-emotion-based norms
for d in all_norms:
    if d[0]==d[1] and d[2]==d[3] and d[4]==d[5] and d[6]==d[7]:
        print("Found non-emotion based norm:", d)
        all_norms.remove(d)
        
print("Without non-emotion based norms, we have:", len(all_norms))
```

```python
# remove emotion-symmetric norms from final list
symmetric_ebsns = []
for d in all_norms:
    inverted = invert_emotion(d)
    if inverted in all_norms:
        print("Emotion inversion of", d, ":", inverted, "already in array, so removing one")
        # choose which one to remove for simplicity
        # prioritize a) d[7]=1 (connect cooperative EP, cooperation with good and a good rep for intuitiveness)
        if d[7]==1 and inverted[7]==0:
            all_norms.remove(inverted)
            print("\tRemoved", inverted)
        elif d[7]==0 and inverted[7]==1:
            all_norms.remove(d)
            print("\tRemoved", d)
        else:
            # THEN prioritize b) d[5]=1 (connect cooperative EP, cooperation with bad and a good rep)
            if d[5]==1 and inverted[5]==0:
                all_norms.remove(inverted)
                print("\tRemoved", inverted)
            elif d[5]==0 and inverted[5]==1:
                all_norms.remove(d)
                print("\tRemoved", d)
            else:
                # Here prioritize d[1]=1 (coonect cooperative EP with punishing bad individuals)
                if d[1]==1 and inverted[1]==0:
                    all_norms.remove(inverted)
                    print("\tRemoved", inverted)
                elif d[1]==0 and inverted[1]==1:
                    all_norms.remove(d)
                    print("\tRemoved", d)
                else:
                    all_norms.remove(d)
                    print("\t\tLast case, removed:", d)


print("Total norms without base and emotion-based symmetrics, with self-symmetrics:", len(all_norms))
```

```python
import numpy as np

def invert(d: list):
    return [1-d[4], 1-d[5], 1-d[6], 1-d[7], 1-d[0], 1-d[1], 1-d[2], 1-d[3]]

def invert_emotion(d):
    return [d[1], d[0], d[3], d[2], d[5], d[4], d[7], d[6]]

# 1. Generate all 256
all_256 = [[int(bit) for bit in format(i, '08b')] for i in range(256)]

# 2. Filter out Non-Emotion Based
eb_norms = [d for d in all_256 if not (d[0]==d[1] and d[2]==d[3] and d[4]==d[5] and d[6]==d[7])]

# 3. Handle Symmetrics (Action Inversion)
unique_after_action = []
processed_action = set()

for d in eb_norms:
    d_tuple = tuple(d)
    inv_tuple = tuple(invert(d))
    if d_tuple not in processed_action and inv_tuple not in processed_action:
        unique_after_action.append(d)
        processed_action.add(d_tuple)
        processed_action.add(inv_tuple)
    elif d_tuple == inv_tuple: # Keep self-symmetrics
        if d_tuple not in processed_action:
            unique_after_action.append(d)
            processed_action.add(d_tuple)

# 4. Handle Emotion Inversion (The part that was failing)
final_norms = []
processed_emotion = set()

for d in unique_after_action:
    d_tuple = tuple(d)
    if d_tuple in processed_emotion:
        continue
        
    inverted = invert_emotion(d)
    inv_tuple = tuple(inverted)
    
    # If the norm is self-symmetric regarding emotion, just add it
    if d_tuple == inv_tuple:
        final_norms.append(d)
        processed_emotion.add(d_tuple)
        continue

    # Priority Logic: Pick the 'best' representative between d and inverted
    # Since inverted might not be in unique_after_action, we check both
    best_rep = d
    
    # Priority sequence: bit 7, then bit 5, then bit 1
    for bit in [7, 5, 1]:
        if d[bit] == 1 and inverted[bit] == 0:
            best_rep = d
            break
        if inverted[bit] == 1 and d[bit] == 0:
            best_rep = inverted
            break
    else:
        # Tie-breaker (lexicographical)
        best_rep = d if d_tuple > inv_tuple else inverted

    final_norms.append(best_rep)
    processed_emotion.add(d_tuple)
    processed_emotion.add(inv_tuple)

print(f"Final unique EBSNs: {len(final_norms)}")
for n in final_norms: print(n)
```
