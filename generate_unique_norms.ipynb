```python
# Online Python Playground
# Use the online IDE to write, edit & run your Python code
# Create, edit & delete files online

def invert(d: list):
    return [1-d[4], 1-d[5], 1-d[6], 1-d[7], 1-d[0], 1-d[1], 1-d[2], 1-d[3]]

def invert_emotion(d):
    return [d[1], d[0], d[3], d[2], d[5], d[4], d[7], d[6]]

all_norms = []

# Generate all
for i in range(256):
    d = [int(bit) for bit in format(i, '08b')]
    all_norms.append(d)
 
# remove symmetrics
for d in all_norms:
    if invert(d) in all_norms:
        print("Inversion of", d, ":", invert(d), "already in array, so removing it")
        all_norms.remove(invert(d))
print("Total norms without symmetrics:", len(all_norms))
```

```python
for norm in all_norms:
    if invert(norm) == invert_emotion(norm):
        print(norm, invert(norm), invert_emotion(norm))
```

```python
# create list of self-symmetrics
self_symmetrics = []
for i in range(256):
    d = [int(bit) for bit in format(i, '08b')]
    if d == invert(d):
        print("Self-symmetric:", d)
        self_symmetrics.append(d)
print("\tThere are", len(self_symmetrics), "self symmetric ebsn's")

# add the self-symmetrics to the list if they aren't there already
for ss in self_symmetrics:
    all_norms.append(ss)


print("Total norms without symmetrics but with self-symmetrics:", len(all_norms))
```

```python
# remove non-emotion-based norms
for d in all_norms:
    if d[0]==d[1] and d[2]==d[3] and d[4]==d[5] and d[6]==d[7]:
        print("Found non-emotion based norm:", d)
        all_norms.remove(d)
        
print("Without non-emotion based norms, we have:", len(all_norms))
```

```python
        
# remove emotion-symmetric norms from final list
symmetric_ebsns = []
for d in all_norms:
    inverted = invert_emotion(d)
    if inverted in all_norms:
        print("Emotion inversion of", d, ":", inverted, "already in array, so removing one")
        # choose which one to remove for simplicity
        # prioritize a) d[7]=1 (connect cooperative EP, cooperation with good and a good rep for intuitiveness)
        if d[7]==1 and inverted[7]==0:
            all_norms.remove(inverted)
            print("\tRemoved", inverted)
        elif d[7]==0 and inverted[7]==1:
            all_norms.remove(d)
            print("\tRemoved", d)
        else:
            # THEN prioritize b) d[5]=1 (connect cooperative EP, cooperation with bad and a good rep)
            if d[5]==1 and inverted[5]==0:
                all_norms.remove(inverted)
                print("\tRemoved", inverted)
            elif d[5]==0 and inverted[5]==1:
                all_norms.remove(d)
                print("\tRemoved", d)
            else:
                print("\tBoth cared not for d[G,C,N] or d[B,C,N], so we remove the first one")

print("Total norms without base and emotion-based symmetrics, with self-symmetrics:", len(all_norms))
```
