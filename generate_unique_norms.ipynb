```python
# Online Python Playground
# Use the online IDE to write, edit & run your Python code
# Create, edit & delete files online

def invert(d: list):
    return [1-d[4], 1-d[5], 1-d[6], 1-d[7], 1-d[0], 1-d[1], 1-d[2], 1-d[3]]


all_norms = []
# Generate all
for i in range(256):
    d = [int(bit) for bit in format(i, '08b')]
    all_norms.append(d)
 
# remove symmetrics
for d in all_norms:
    if invert(d) in all_norms:
        print("Inversion of", d, ":", invert(d), "already in array, so removing it")
        all_norms.remove(invert(d))
print("Total norms so far:", len(all_norms))

# add self-symmetric
self_symmetrics = []
for i in range(256):
    d = [int(bit) for bit in format(i, '08b')]
    if d == invert(d):
        print("Self-symmetric:", d, "and", invert(d))
        self_symmetrics.append(d)
print("There are", len(self_symmetrics), "self symmetric ebsn's")

for ss in self_symmetrics:
    if ss in all_norms:
        print(ss, "already in the list")
    else:
        all_norms.append(ss)

print("Total norms so far:", len(all_norms))

symmetric_ebsns = []
for d in all_norms:
    if d[0]==1-d[1] and d[2]==1-d[3] and d[4]==1-d[5] and d[6]==1-d[7]:
        symmetric_ebsns.append(d)
        print("EBSN self-symmetric", d)
        all_norms.remove(d)

print("Number of all unique norms", len(all_norms))
print(all_norms)

```
